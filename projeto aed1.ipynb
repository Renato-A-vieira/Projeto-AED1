```python
projeto_aed1 = "Projeto AED1"
print("\n" + "=" * len(projeto_aed1.center(80)))
print("\33[1m" + projeto_aed1.center(80) + "\33[0m")
print("=" * len(projeto_aed1.center(80)))


import random
import string

usuarios = []

def validar_nome(nome):
    return len(nome) >= 3

def validar_email(email):
    return len(email) >= 8 or "@" in email or ".com" in email

def validar_telefone(telefone):
    return telefone.isdigit() and len(telefone) == 11

""" È a validação do nome, email e telefone, ta bem basico esse projeto"""

def gerar_login(usuarios):
    logins_criados = {u["login"] for u in usuarios}
    while True:
        login = ''.join(random.choice(string.digits) for _ in range(8))#''.join é usado para juntar a palavra, quando é gerado é gerado assim: 1 , 2 , 3.
        if login not in logins_criados:
            return login

def gerar_senha():
    senha_criada = {s["senha"] for s in usuarios}
    caracteres = string.ascii_letters + string.digits #criado para juntar letras e numeros
    while True:
        senha = ''.join(random.choice(caracteres) for _ in range(12))
        if senha not in senha_criada:
            return senha
""" Aqui é geração aleatoria do login e senha """

def cadastrar_cliente():
    nome = input("Digite seu nome: ").strip() #strip é usado para limpar os espaços no final da frase.
    while len(nome) < 3:
        print(" O nome deve ter pelo menos 3 caracteres.")
        nome = input("Digite novamente: ")
    email = input("Digite seu e-mail: ").strip()
    while len(email) < 5 or "@" not in email or ".com" not in email:
        print("E-mail inválido. Exemplo: nome@fatec.com")
        email = input("Digite novamente: ")
    telefone = input("Digite seu telefone com o DDD sem o 0:")#não tem strip porque tem limite de numeros.
    while not telefone.isdigit():
        print("Utilize apenas números.")
        telefone = input("Digite novamente.")
    
    while len(telefone) != 11:
        print("Número de telefone errado.")
        telefone= input("Digite novamente.")
        while not telefone.isdigit():
            print("Utilize apenas números.")
            telefone = input("Digite novamente.")
    telefone = int(telefone)

    login = gerar_login(usuarios)
    senha = gerar_senha()
    
    clientes = {"nome" : nome,
             "telefone" : telefone,
             "email" : email,
             "login" : login,
             "senha" : senha
}

    usuarios.append(clientes)
            
    print ("\n" + "=" * 40)
    print(f"{"FICHA DE CADASTRO":^40}")
    
    print(f"{'\nNome:':<12}{clientes["nome"]}")
    print(f"{'E-mail:':<12}{clientes["telefone"]}")
    print(f"{'Telefone:':<12}{clientes["email"]}")
    print(f"{'Login:':<12}{clientes["login"]}")
    print(f"{'Senha:':<12}{clientes["senha"]}")
    print ("=" * 40)

""" Aqui é a criação do cadastro """

def painel():
    while True: 
        print("\n" + "=" * 40)
        print(f"{'MENU PRINCIPAL':^40}")
        print("=" * 40)
        print("1 - Cadastrar Cliente")
        print("2 - Sair")
        print("=" * 40)

        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            cadastrar_cliente()
            break
        elif opcao == "2":
            print("\nEncerrando o programa... ")
            break  
        else:
            input("Opção inválida! Pressione Enter para tentar novamente.")


painel()
```
