```python
projeto_aed1 = "Projeto AED1"
print("\n" + "=" * len(projeto_aed1.center(80)))
print("\33[1m" + projeto_aed1.center(80) + "\33[0m")
print("=" * len(projeto_aed1.center(80)))


import random
import string

usuarios = []   #lista onde todos os clientes cadastrados serão armazenados

def gerar_login():
    logins_criados = {l["login"] for l in usuarios}  # Compara logins já criados para evitar duplicação.
    while True:
        login = ''.join(random.choice(string.digits) for _ in range(8))#''.join é usado para juntar a palavra, quando é gerado é gerado assim: 1 , 2 , 3.
        if login not in logins_criados:
            return login

def gerar_senha():
    senha_criada = {s["senha"] for s in usuarios}
    caracteres = string.ascii_letters + string.digits #criado para juntar letras e numeros
    while True:
        senha = ''.join(random.choice(caracteres) for _ in range(12))
        if senha not in senha_criada:
            return senha
""" Aqui é geração aleatoria do login e senha """

def cadastrar_cliente():
    nome = input("Digite seu nome: ").strip() #strip é usado para limpar os espaços no final da frase.
    while len(nome) < 3:
        print(" O nome deve ter pelo menos 3 caracteres.")
        nome = input("Digite novamente: ")
    email = input("Digite seu e-mail: ").strip()
    while len(email) < 8 or "@" not in email or ".com" not in email:
        print("E-mail inválido. Exemplo: nome@fatec.com")
        email = input("Digite novamente: ")
    telefone = input("Digite seu telefone com o DDD sem o 0:")#não tem strip porque tem limite de numeros.
    while not telefone.isdigit():
        print("Utilize apenas números.")
        telefone = input("Digite novamente.")
    
    while len(telefone) != 11:
        print("Número de telefone errado. ex : 01234567890")
        telefone= input("Digite novamente.")
        while not telefone.isdigit():
            print("Utilize apenas números.")
            telefone = input("Digite novamente.")
    telefone = int(telefone)

    login = gerar_login()
    senha = gerar_senha()
    
    clientes = {"nome" : nome,
             "telefone" : telefone,
             "email" : email,
             "login" : login,
             "senha" : senha
}

    usuarios.append(clientes)

    print ("\n" + "=" * 40)
    print(f"{"FICHA DE CADASTRO":^40}")
    print(f"{'\nNome:':<12}{clientes["nome"]}")
    print(f"{'E-mail:':<12}{clientes["email"]}")
    print(f"{'Telefone:':<12}{clientes["telefone"]}")
    print(f"{'Login:':<12}{clientes["login"]}")
    print(f"{'Senha:':<12}{clientes["senha"]}")
    print ("=" * 40)


def calcular_juros(valor, taxa, tempo):
    montante = valor * ((1 + taxa/100) ** tempo)
    return montante


def menu_investimento(usuario):

    while True:
        print("\n" + "=" * 40)
        print(f"{'MENU DE INVESTIMENTO':^40}")
        print("=" * 40)
        print("1 - Fazer novo investimento")
        print("2 - Verificar seu investimento")
        print("3 - Sair")

        opcao = input("Digite sua opção: ").strip()
        if opcao == "1":        
            while True:
                valor_str = input("Digite o valor a investir em reais: ").replace(",", ".")
                if valor_str.replace('.', '', 1).isdigit():
                    valor = float(valor_str)
                    break
                print("Digite apenas números! Exemplo: 100 ou 100.50")

            while True:
                taxa_str = input("Digite a taxa de juros (% ao mês): ").replace(",", ".")
                if taxa_str.replace('.', '', 1).isdigit():
                    taxa = float(taxa_str)
                    break
                print("Digite apenas números!")

            while True:
                tempo_str = input("Digite o período em meses: ").strip()
                if tempo_str.isdigit():
                    tempo = int(tempo_str)
                    break
                print("Digite um número válido.")

            montante = calcular_juros(valor, taxa, tempo)
            ganho = montante - valor

            usuario["valor"] = valor
            usuario["taxa"] = taxa
            usuario["tempo"] = tempo
            usuario["montante"] = montante
            usuario["ganho"] = ganho

            print(f"\nValor final após {tempo} meses: R$ {montante:.2f}")
            print(f"Lucro obtido: R$ {ganho:.2f}")

            
        elif opcao == "2":
            if "valor" in usuario:
                print(f"O valor investido foi : {usuario['valor']:.2f}")
                print(f"O retorno esperado : {usuario['ganho']:.2f}")
                print(f"Total no final dos {usuario['tempo']} meses: R${usuario['montante']:.2f}")
                break
            else:
                print("\nVocê ainda não fez nenhum investimento.")
                print("O valor investido foi : R$ 0.00")
                print("O retorno esperado : R$ 0.00")
                print("Total no final dos 0 meses: R$ 0.00")
                continue

                
        elif opcao == "3":
            print("Muito obrigado e tenha um otimo dia")
            break

        else:
            print("Digitou uma opção invalida, digite novamente")
            return False
            
def painel():
    while True: 
        print("\n" + "=" * 40)
        print(f"{'MENU PRINCIPAL':^40}")
        print("=" * 40)
        print("1 - Cadastrar Cliente")
        print("2 - Sair")
        print("=" * 40)

        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            cadastrar_cliente()
            break
        elif opcao == "2":
            print("\nEncerrando o programa... ")
            break  
        else:
            input("Opção inválida! Pressione Enter para tentar novamente.")

def login_usuario():
    print("\n" + "="*40)
    print(f"\n{"Tela do Login ":^12}")
    login = input("Digite seu login").strip()
    senha = input("Digite sua senha").strip()

    for usuario in usuarios:
        if usuario["login"] == login and usuario["senha"] == senha:
            print(f"Bem-vindo {usuario['nome']}, login bem-sucedido!")
            if "investimentos" not in usuario:
                usuario["investimentos"] = 0.0            
            menu_investimento(usuario) #função valdia do menu de investimento
            return True
    print("Login/senha errada")
    return False

painel()
```
